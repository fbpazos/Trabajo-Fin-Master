```python
import numpy as np
from BNumMet.NonLinear import secant, bisect, newton, zBrentDekker, IQI
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
fun = lambda x: x**2 - 2
funPrime = lambda x: 2 * x
interval = [1, 2]

output = [
    bisect(fun, interval, iters=True),
    secant(fun, interval, iters=True),
    newton(fun, funPrime, interval[0], iters=True),
    IQI(fun, [1, 1.5, 2], iters=True),
    zBrentDekker(fun, interval, iters=True),
]

display(output)
```

```python
f = lambda x: sp.jv(0, x)
interval = lambda n: [n * np.pi, (n + 1) * np.pi]

zeros = []
for n in range(0, 10):
    zeros.append(zBrentDekker(f, interval(n)))

x = np.arange(1, 10 * np.pi, np.pi / 50)
y = f(x)

plt.plot(x, y)
plt.plot(zeros, np.zeros(len(zeros)), "ro")
plt.axhline(0, color="k")
plt.show()
```
